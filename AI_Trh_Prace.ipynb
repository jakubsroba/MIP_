```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import random

# pekné grafy
plt.rcParams["figure.figsize"] = (8,5)
plt.rcParams["axes.grid"] = True

# ============================
# 1. GENEROVANIE JOB DÁT
# ============================

random.seed(42)
np.random.seed(42)

countries = ["Slovakia", "Germany", "Netherlands"]
years = [2023, 2024, 2025]

records = []

for year in years:
    for country in countries:
        # Koľko pracovných ponúk má krajina v danom roku
        if country == "Germany":
            total_jobs = np.random.randint(250, 400)
        elif country == "Netherlands":
            total_jobs = np.random.randint(220, 330)
        else:  # Slovakia
            total_jobs = np.random.randint(120, 200)

        # Základný podiel AI jobov podľa krajiny
        if country == "Germany":
            base_ai_share = 0.28   # Nemecko - dosť AI
        elif country == "Netherlands":
            base_ai_share = 0.35   # Holandsko - high AI adopcia
        else:  # Slovakia
            base_ai_share = 0.15   # Slovensko - ešte nízke, ale rastie

        # Postupný nárast AI z roka na rok
        growth_factor = 1 + 0.05*(year-2023)  # 2023=1.00, 2024=1.05, 2025=1.10
        ai_share = base_ai_share * growth_factor

        ai_jobs = int(total_jobs * ai_share)
        non_ai_jobs = total_jobs - ai_jobs

        # AI job titles = práce, ktoré AI vytvára
        ai_titles = [
            "AI Engineer",
            "Machine Learning Engineer",
            "Data Scientist (NLP)",
            "Computer Vision Specialist",
            "AI Automation Specialist",
            "LLM Integration Engineer",
            "MLOps Engineer",
        ]

        # non-AI job titles = práce, ktoré sú skôr rutinné
        # = tie, ktoré sú ohrozené automatizáciou
        non_ai_titles = [
            "Office Administrator",
            "Clerical Assistant",
            "Back Office Support",
            "Warehouse Operator",
            "Production Line Operator",
            "Logistics Coordinator",
            "Accounting Assistant",
            "Customer Service Specialist",
        ]

        # generuj AI joby
        for i in range(ai_jobs):
            if country == "Slovakia":
                salary = np.random.randint(2200, 3200)
            elif country == "Germany":
                salary = np.random.randint(4000, 6500)
            else:  # Netherlands
                salary = np.random.randint(4200, 6800)

            records.append({
                "year": year,
                "country": country,
                "title": random.choice(ai_titles),
                "salary": salary,
                "is_AI": True
            })

        # generuj non-AI joby
        for i in range(non_ai_jobs):
            if country == "Slovakia":
                salary = np.random.randint(1100, 2000)
            elif country == "Germany":
                salary = np.random.randint(2000, 3500)
            else:  # Netherlands
                salary = np.random.randint(2100, 3600)

            records.append({
                "year": year,
                "country": country,
                "title": random.choice(non_ai_titles),
                "salary": salary,
                "is_AI": False
            })

adzuna_df = pd.DataFrame(records)

print("Náhľad pracovných ponúk (simulované dáta z pracovného portálu):")
display(adzuna_df.head())

print("Počet celkových ponúk v datasete:", len(adzuna_df))


# ============================
# 2. MAKRO DÁTA (Eurostat / OECD)
# ============================
# ict_specialists_share_pct = koľko % pracovnej sily tvoria IT/ICT špecialisti
# automation_risk_pct = aké % pracovných miest je ohrozených automatizáciou
# tieto hodnoty simulujeme tak, aby dávali zmysel (Slovensko = viac rizika)

eurostat_data = [
    {"year":2023, "country":"Slovakia",    "ict_specialists_share_pct": 4.1, "employment_rate_pct": 68.5},
    {"year":2023, "country":"Germany",     "ict_specialists_share_pct": 5.8, "employment_rate_pct": 77.2},
    {"year":2023, "country":"Netherlands", "ict_specialists_share_pct": 6.2, "employment_rate_pct": 79.1},

    {"year":2024, "country":"Slovakia",    "ict_specialists_share_pct": 4.5, "employment_rate_pct": 69.0},
    {"year":2024, "country":"Germany",     "ict_specialists_share_pct": 6.1, "employment_rate_pct": 77.5},
    {"year":2024, "country":"Netherlands", "ict_specialists_share_pct": 6.6, "employment_rate_pct": 79.4},

    {"year":2025, "country":"Slovakia",    "ict_specialists_share_pct": 4.9, "employment_rate_pct": 69.4},
    {"year":2025, "country":"Germany",     "ict_specialists_share_pct": 6.5, "employment_rate_pct": 77.8},
    {"year":2025, "country":"Netherlands", "ict_specialists_share_pct": 7.0, "employment_rate_pct": 79.7},
]

oecd_data = [
    {"year":2023, "country":"Slovakia",    "automation_risk_pct": 33.0},
    {"year":2023, "country":"Germany",     "automation_risk_pct": 27.5},
    {"year":2023, "country":"Netherlands", "automation_risk_pct": 25.0},

    {"year":2024, "country":"Slovakia",    "automation_risk_pct": 32.2},
    {"year":2024, "country":"Germany",     "automation_risk_pct": 26.9},
    {"year":2024, "country":"Netherlands", "automation_risk_pct": 24.6},

    {"year":2025, "country":"Slovakia",    "automation_risk_pct": 31.5},
    {"year":2025, "country":"Germany",     "automation_risk_pct": 26.1},
    {"year":2025, "country":"Netherlands", "automation_risk_pct": 24.0},
]

eurostat_df = pd.DataFrame(eurostat_data)
oecd_df = pd.DataFrame(oecd_data)

print("\nNáhľad makrodát (Eurostat):")
display(eurostat_df.head())

print("\nNáhľad makrodát (OECD):")
display(oecd_df.head())


# ============================
# 3. AGREGÁCIA JOBOV
# ============================

# koľko AI jobov a non-AI jobov je v každej krajine v každom roku
ai_counts = (
    adzuna_df
    .groupby(["year", "country", "is_AI"])
    .agg(
        job_count=("title","count"),
        avg_salary=("salary","mean")
    )
    .reset_index()
)

print("\nPočet AI / non-AI ponúk podľa roka, krajiny a typu pozície:")
display(ai_counts.head(12))

# koľko AI jobov má krajina v danom roku
ai_only_counts = (
    adzuna_df[adzuna_df["is_AI"] == True]
    .groupby(["year", "country"])
    .agg(
        ai_job_count=("title","count"),
        ai_avg_salary=("salary","mean")
    )
    .reset_index()
)

print("\nAI pozície podľa roka/krajiny (počet a priemerný plat):")
display(ai_only_counts.head(12))

# koľko všetkých jobov má krajina v danom roku
total_counts = (
    adzuna_df
    .groupby(["year", "country"])
    .agg(
        total_job_count=("title","count"),
        total_avg_salary=("salary","mean")
    )
    .reset_index()
)

# spojíme total + AI, vyrátame podiel AI
combined = pd.merge(total_counts, ai_only_counts, on=["year","country"], how="left")

combined["ai_job_count"]   = combined["ai_job_count"].fillna(0)
combined["ai_avg_salary"]  = combined["ai_avg_salary"].fillna(0)

combined["ai_share_pct"] = (
    combined["ai_job_count"] / combined["total_job_count"] * 100
)

print("\nZhrnutie za rok/krajinu: veľkosť trhu práce, AI joby, podiel AI (%):")
display(combined)


# ============================
# 4. TREND AI POZÍCIÍ (spolu všetky krajiny)
# ============================

ai_trend_total = (
    adzuna_df[adzuna_df["is_AI"] == True]
    .groupby("year")
    .agg(ai_job_count_total=("title","count"))
    .reset_index()
)

ai_trend_total["pct_change_vs_prev_year"] = (
    ai_trend_total["ai_job_count_total"].pct_change() * 100
)

print("\nTrend AI pracovných ponúk (celý súbor spolu):")
display(ai_trend_total)


# ============================
# 5. SPOJENIE S MAKRODÁTAMI
# ============================

merged_level1 = pd.merge(combined, eurostat_df,
                         on=["year","country"], how="left")

merged_full = pd.merge(merged_level1, oecd_df,
                       on=["year","country"], how="left")

print("\nVšetko pokope (krajina+rok): AI podiel, platy, IT podiel, riziko automatizácie:")
display(merged_full)


# ============================
# 6. VIZUALIZÁCIE
# ============================

# (A) Koláčový graf: AI vs non-AI v celej vzorke (2023-2025 spolu)
ai_total = adzuna_df[adzuna_df["is_AI"]==True].shape[0]
non_ai_total = adzuna_df[adzuna_df["is_AI"]==False].shape[0]

plt.figure(figsize=(5,5))
plt.pie(
    [ai_total, non_ai_total],
    labels=["AI pozície (nové joby)", "Rutinné/klasické pozície"],
    autopct="%1.1f%%",
    startangle=90
)
plt.title("Štruktúra trhu práce vo vzorke")
plt.tight_layout()
plt.show()

# (B) Stĺpcový graf: počet AI pracovných ponúk podľa krajiny a roku
pivot_ai = ai_only_counts.pivot(index="year",
                                columns="country",
                                values="ai_job_count")

pivot_ai.plot(kind="bar")
plt.title("Počet AI pracovných ponúk podľa krajiny a roku")
plt.xlabel("Rok")
plt.ylabel("Počet AI ponúk")
plt.legend(title="Krajina")
plt.tight_layout()
plt.show()

# (C) Čiarový graf: rast AI pracovných ponúk v čase (všetky krajiny spolu)
plt.plot(ai_trend_total["year"],
         ai_trend_total["ai_job_count_total"],
         marker="o")
plt.title("Trend AI pracovných ponúk (všetky krajiny spolu)")
plt.xlabel("Rok")
plt.ylabel("Počet AI ponúk (spolu)")
plt.tight_layout()
plt.show()

# (D) Stĺpcový graf: podiel AI (%) v každej krajine/roku
pivot_share = combined.pivot(index="year",
                             columns="country",
                             values="ai_share_pct")

pivot_share.plot(kind="bar")
plt.title("Podiel AI pozícií na všetkých ponukách (%)")
plt.xlabel("Rok")
plt.ylabel("Podiel AI ponúk (%)")
plt.legend(title="Krajina")
plt.tight_layout()
plt.show()

# (E) Scatter: riziko automatizácie vs podiel AI pozícií
plt.figure(figsize=(8,5))

sizes = merged_full["total_job_count"] * 0.5  # veľkosť bubliny = veľkosť trhu práce

plt.scatter(
    merged_full["automation_risk_pct"],
    merged_full["ai_share_pct"],
    s=sizes,
    alpha=0.6,
    edgecolor="black"
)

for _, row in merged_full.iterrows():
    label = f'{row["country"]} {row["year"]}'
    plt.text(
        row["automation_risk_pct"] + 0.2,
        row["ai_share_pct"] + 0.2,
        label,
        fontsize=8
    )

plt.title("Riziko automatizácie vs. podiel AI ponúk")
plt.xlabel("Riziko automatizácie (%)")
plt.ylabel("Podiel AI ponúk na trhu práce (%)")
plt.grid(True)
plt.tight_layout()
plt.show()

# (F) Čiarový graf: vývoj rizika automatizácie podľa krajiny
automation_trend = oecd_df.pivot(index="year",
                                 columns="country",
                                 values="automation_risk_pct")

automation_trend.plot(marker="o")
plt.title("Vývoj rizika automatizácie práce podľa krajiny")
plt.xlabel("Rok")
plt.ylabel("Riziko automatizácie (%)")
plt.grid(True)
plt.tight_layout()
plt.show()


# ============================
# 7. INTERPRETÁCIA (text pre prezentáciu / záver)
# ============================

def interpretacia(ai_trend_df, merged_df, ai_total, non_ai_total):
    posledny_rok = ai_trend_df["year"].max()
    pocet_posledny = ai_trend_df.loc[
        ai_trend_df["year"] == posledny_rok,
        "ai_job_count_total"
    ].iloc[0]

    # krajina s najvyšším podielom AI jobov
    top_row = merged_df.sort_values("ai_share_pct", ascending=False).iloc[0]
    top_country = top_row["country"]
    top_share = round(top_row["ai_share_pct"], 1)
    top_year = top_row["year"]

    # krajina s najvyšším rizikom automatizácie
    risk_row = merged_df.sort_values("automation_risk_pct", ascending=False).iloc[0]
    risk_country = risk_row["country"]
    risk_val = round(risk_row["automation_risk_pct"], 1)
    risk_year = risk_row["year"]

    print("---- Zhrnutie pre prezentáciu ----")
    print(f"Analyzovali sme približne {ai_total + non_ai_total} pracovných ponúk zo simulovaného open datasetu.")
    print(f"AI pozície tvoria {round(ai_total/(ai_total+non_ai_total)*100,1)} % všetkých ponúk.")
    print(f"Počet AI pracovných ponúk rastie v čase a v roku {posledny_rok} dosahuje {pocet_posledny} ponúk (spolu všetky krajiny).")
    print(f"Najvyšší podiel AI pozícií má {top_country} v roku {top_year} (približne {top_share} % všetkých ponúk).")
    print(f"Najvyššie riziko automatizácie má {risk_country} v roku {risk_year} (asi {risk_val} % ohrozených miest).")
    print("Vidíme teda napätie medzi dvomi javmi:")
    print("  1) staré, rutinné role môžu byť nahradené technológiou,")
    print("  2) nové digitálne role v oblasti AI vznikajú a rastú.")
    print("Krajiny s vyšším podielom IT špecialistov a silnejšou digitalizáciou majú vyšší podiel AI ponúk a zároveň nižšie riziko automatizácie.")
    print("----------------------------------")

interpretacia(ai_trend_total, merged_full, ai_total, non_ai_total)

```
